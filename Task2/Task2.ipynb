```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.optimize import minimize
import main
```

```python
dispersion = 3
a_true = np.array([2, 1, 3])
m = 200.0
t = np.array([i*10.0/m for i in range(int(m))])
y_true = main.function(t, a_true)
y_rand = main.gen_data(dispersion, y_true)
main.find_errors(y_rand)
```

```python
dispersion = 0.4
y_rand = main.gen_data(dispersion, y_true)
main.find_errors(y_rand)
```

Видим, что в обоих случаях с восстановлением коеффициентов хуже всех справляется оранжевая линия - это метод, основанный на максимальном модуле ошибки. Красную линию (сумма квадратов) и зелёную (сумма модулей) сложно отличить друг от друга при малом шуме.

```python
dispersion = 0.2
y_rand = main.gen_data(dispersion, y_true)
plt.ylim(-0, 10)
main.find_errors(y_rand)
```

Посмотрим, как методы среагируют на выбросы.

```python
dispersion = 0.2
y_rand = main.gen_data(dispersion, y_true)
y_rand[-1] = y_rand[1]+50

main.find_errors(y_rand)
```

Видно, что третий метод сильнее всего реагирует на выбросы, далее сравним первые два.

```python
dispersion = 0.2
y_rand = main.gen_data(dispersion, y_true)
y_rand[10] = y_rand[1]+10
y_rand[20] = y_rand[1]+ 6
y_rand[10] = y_rand[1]+10
y_rand[40] = y_rand[1]+ 7
y_rand[10] = y_rand[1]+10
y_rand[30] = y_rand[1]+ 6
y_rand[50] = y_rand[1]+10
y_rand[60] = y_rand[1]+ 7
main.find_errors(y_rand)
```
