```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.optimize import minimize
import main
```

```python
dispersion = 3
a_true = np.array([2, 1, 3])
m = 200.0
t = np.array([i*10.0/m for i in range(int(m))])
y_true = main.function(t, a_true)
y_rand = main.gen_data(dispersion, y_true)
main.find_errors(y_rand)
```
